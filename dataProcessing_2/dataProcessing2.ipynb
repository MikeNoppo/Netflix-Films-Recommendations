# Penggabungan Data

Pada tahap ini, kita akan menggabungkan dataset `new_metadata` dengan dataset `credits` berdasarkan kolom `'id'`. Penggabungan ini dilakukan menggunakan fungsi `pd.merge()`, yang akan menggabungkan kedua dataset tersebut pada baris-baris yang memiliki nilai `'id'` yang sama.

```python
data = pd.merge(new_metadata, credits, on='id')
```

```python
data
```

```python
# Kita akan mengonversi kolom "genres", "cast", dan "crew" menjadi sebuah list. Jika kita perhatikan dengan seksama, 
# kita akan menyadari bahwa kolom-kolom tersebut sebenarnya adalah string yang merepresentasikan list yang berisi 
# dictionary.
import ast
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))  # Fungsi "literal_eval" membantu mengubah string menjadi objek Python asli (dalam hal ini, list of dictionaries)
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

# Ekstraksi Genre Film

Pada tahap ini, kita akan mengambil genre film secara individual, seperti **adventure**, **action**, dan untuk genre Science fiction kita ubah penulisan nya menjadi **sci-fi** di dataset. tetapi di website nanti yang akan di tampilkan tetap Science fiction , dan sebagainya, menggunakan fungsi berikut:

```python
def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.nan
    else:
        return (st.join(gen))
```

```python
# implementasi fungsi make_genreList
data['genres_list'] = data['genres'].map(lambda x: make_genresList(x))
```

```python
data['genres_list']
```

```python
# lakukan hal yang sama untuk cast
# mengambil nama aktor dari daftar cast , jika tidak ada actor yang tersedia kembalikan NaN
# actor 1
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.nan # jika daftar cast kosong Kembalikan NaN
    else:
        return (casts[0])
    
# actor 2
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.nan # jika daftar cast kosong Kembalikan NaN
    else:
        return (casts[1])   
    
# actor 3
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.nan # jika daftar cast kosong Kembalikan NaN
    else:
        return (casts[2])
```

```python
# terapkan pada kolom cast untuk aktor pertama, kedua, dan ketiga
data['actor_1_name'] = data['cast'].map(lambda x: get_actor1(x))
data['actor_2_name'] = data['cast'].map(lambda x: get_actor2(x))
data['actor_3_name'] = data['cast'].map(lambda x: get_actor3(x))
```

```python
data[['actor_1_name', 'actor_2_name', 'actor_3_name']]
```

```python
# lakukan hal yang sama untuk director
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.nan
    else:
        return (st.join(dt))
```

```python
data['director_name'] = data['crew'].map(lambda x: get_directors(x))
```

```python
data['director_name'] 
```

# Pengecekan Duplikasi Data

Setelah penggabungan data, kami memeriksa apakah ada baris yang terduplikasi. Kami menggunakan fungsi `duplicated()` dari Pandas untuk mengidentifikasi baris-baris yang terduplikasi. Berikut ini adalah hasil dari pengecekan duplikasi, dengan menampilkan semua baris yang memiliki duplikasi.

```python
duplicates = new[new.duplicated(keep=False)]  # keep=False menampilkan semua duplikat, termasuk yang pertama
duplicates_with_index = duplicates.reset_index()  # Menyimpan indeks asli untuk referensi
display(duplicates_with_index)
```

# Menghapus Duplikasi Data Berdasarkan `movie_title`

Pada langkah ini, kita akan menghapus data yang duplikat berdasarkan kolom **`movie_title`**. Kenapa kita menggunakan **`movie_title`** sebagai acuan? Karena judul film adalah informasi unik yang merepresentasikan setiap film, sehingga memungkinkan kita untuk mengidentifikasi duplikasi dengan lebih akurat.

Jika kita menghapus duplikasi berdasarkan **`index`**, hal ini tidak efektif dalam menghilangkan duplikasi yang sebenarnya karena **index** dapat berbeda antara data lama (`old`) dan data baru (`new`). Sebagai contoh, pada output sebelumnya, kita melihat bahwa data duplikat pada judul film **"a dog's breakfast"** terdapat di dua baris dengan **index** yang berbeda:
- Data lama (`old`) berada di **index 0**.
- Data baru (`new`) berada di **index 6**.

Menggunakan **`index`** dalam hal ini akan mengabaikan fakta bahwa kedua baris tersebut sebenarnya merujuk pada film yang sama. Oleh karena itu, kita menggunakan **`movie_title`** untuk menghapus duplikat, memastikan hanya satu baris yang tersisa untuk setiap film.


```python
new.drop_duplicates(subset="movie_title", keep='last', inplace=True) 
```

```python
duplicates = new[new.duplicated()]
print(f"Jumlah duplikasi setelah penghapusan: {duplicates.shape[0]}")
duplicates
```
